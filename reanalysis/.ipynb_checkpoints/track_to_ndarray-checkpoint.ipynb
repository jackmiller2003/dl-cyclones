```python
import json
with open('../tracks/proc_tracks.json', 'r') as f:
    tracks = json.load(f)
print(tracks[list(tracks.keys())[0]])
```

```python
import dask
cyclone_thunks = []
for ssid in list(tracks.keys())[2000:2010]:
    track = tracks[ssid]
    cyclone_thunks.append(dask.delayed(
        lib.track_to_ndarray(track['iso_times'], [(lat, long) for [long, lat] in track['coordinates']])
    ))
cyclones = dask.compute(*cyclone_thunks)
```
